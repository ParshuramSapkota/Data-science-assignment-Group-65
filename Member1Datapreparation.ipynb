```python
# ===============================
# Part 1: Data Preparation & Descriptive Statistics
# ===============================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
df1 = pd.read_csv("dataset1.csv")
df2 = pd.read_csv("dataset2.csv")

# Quick check
print("Dataset 1 Shape:", df1.shape)
print("Dataset 2 Shape:", df2.shape)

# Handle missing values (simple drop for now, or fill if needed)
df1 = df1.dropna()
df2 = df2.dropna()

# ----- Quantitative Variables from dataset1 -----
numeric_cols = ["bat_landing_to_food", "seconds_after_rat_arrival", "hours_after_sunset"]

for col in numeric_cols:
    print(f"\n--- {col} ---")
    print("Mean:", df1[col].mean())
    print("Median:", df1[col].median())
    print("Mode:", df1[col].mode()[0])
    print("Std Dev:", df1[col].std())
    print("Variance:", df1[col].var())
    print("Range:", df1[col].max() - df1[col].min())
    print("IQR:", df1[col].quantile(0.75) - df1[col].quantile(0.25))
    print("Skewness:", df1[col].skew())

# ----- Visualization -----
plt.figure(figsize=(8,5))
sns.histplot(df1["bat_landing_to_food"], kde=True, bins=20)
plt.title("Histogram of Bat Landing to Food")
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x=df1["bat_landing_to_food"])
plt.title("Boxplot of Bat Landing to Food")
plt.show()
```
